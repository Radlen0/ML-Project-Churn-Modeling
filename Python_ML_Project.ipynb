```python
# Imports
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, PolynomialFeatures
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
```

# **Chargement des données.**



```python
# chargement des donnée
df = pd.read_csv('Churn_Modelling.csv')

# afficher les 5 premier lignes des donnée
display(df.head(5))
```

# **Analyse des propriétés des données**

```python
# Affiche le nombre total d'éléments (lignes * colonnes)
print(f"Nombre total de cellules : {df.size}")

# Affiche le format (lignes, colonnes)
print(f"Format du dataset (Lignes, Colonnes) : {df.shape}\n")

print("Résumé technique des colonnes :")
df.info()

# Affiche les statistiques descriptives pour les colonnes numériques
print("\nStatistiques descriptives :")
display(df.describe())
```

# **Clean up & Séparation des données**

```python
# On retire les identifiants qui n'ont aucune valeur prédictive
df = df.drop(columns=['CustomerId'])

# X = toutes les colonnes SAUF la cible 'Exited'
x = df.drop(columns=['Exited'])
# y = la colonne que l'on veut prédire
y = df['Exited']


# Split des données (Entraînement / Test)
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)

# Initialisation du scaler (normalization)
scaler = StandardScaler()

# Transformation du set d'entraînement et de test
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

# **Apprentissage & évaluation des modèles.**



*   **La régression logistique**
*   **Random Forest**
*   **SVM (Support Vector Machine)**
*   **XGBoost**
*   **Réseau de Neurones**



```python
models = {
    "Régression Logistique": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1),
    "SVM (Support Vector Machine)": SVC(probability=True),
    "XGBoost": XGBClassifier(),
    "Réseau de Neurones": MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500)
}

for name, model in models.items():
    # Entraînement
    model.fit(x_train_scaled, y_train)

    # Prédiction
    y_pred = model.predict(x_test_scaled)

    # Calcul du score
    score = accuracy_score(y_test, y_pred)

    print(f"{name} : {score:.4f}")
```

# **Optimisations**



*   **Feature Engineering**



```python
# Création des caractéristiques polynomiales
poly = PolynomialFeatures(degree=3, include_bias=False)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

# Standardisation
scaler = StandardScaler()
x_train_poly_scaled = scaler.fit_transform(x_train_poly)
x_test_poly_scaled = scaler.transform(x_test_poly)

# Modèle
model = LogisticRegression(max_iter=5000, random_state=42)
model.fit(x_train_poly_scaled, y_train)

# Évaluation
y_pred = model.predict(x_test_poly_scaled)
print(f"Score nouveau (Régression Logistique) : {accuracy_score(y_test, y_pred):.4f}")
```

*  **Hyperparamétrisation**

```python

from sklearn.model_selection import GridSearchCV

# Définition de la grille de paramètres
param_grid = {
    'n_estimators': [100, 250, 500, 750], 
    'max_depth': [3, 4, 6, 8, 10, 12], 
    'learning_rate': [0.01, 0.05, 0.1, 0.15, 0.2, 0.3], 
    'subsample': [0.6, 0.7, 0.8, 0.9, 1.0], 
    'colsample_bytree': [0.6, 0.7, 0.8, 0.9, 1.0],
    'min_child_weight': [1, 3, 5, 7]
}

# Configuration du GridSearch
grid_search = GridSearchCV(
    estimator=XGBClassifier(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

# Entraînement
grid_search.fit(x_train_scaled, y_train)

# Affichage des meilleurs résultats
print(f"Meilleurs paramètres : {grid_search.best_params_}")
print(f"Meilleur score en CV : {grid_search.best_score_:.4f}")
```
